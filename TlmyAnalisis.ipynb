```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./resources/tlmy.csv')
```

```python

df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df["packetNumber"] = df["packetNumber"].astype(int)
df["OBCUpTime"] = df["OBCUpTime"].astype(int)
df["TempBatt"] = df["TempBatt"].astype(int)
df["frameType"] = df["frameType"].astype(int)
df["UTCTime"] = df["UTCTime"].astype(int)
df["dt"] = pd.to_datetime(df["UTCTime"], unit='s')
```

```python
df.sort_values(by=['UTCTime'], inplace=True)
#df.set_index('packetNumber', inplace=True)
```

```python
df = df.drop_duplicates(subset=['UTCTime'])
df.sort_values(by=['UTCTime'], inplace=True)
```

```python
plt.figure(figsize=(20,10))
dfr = df.iloc[6000:6100, :]
plt.plot(dfr['OBCUpTime'], dfr['TXcurrent'])
plt.plot(dfr['OBCUpTime'], dfr['RXcurrent'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
dfr = df.iloc[500:800, :]
cols = ['obcT1', 'TempBatt', 'IMTQ_temp', 'tempP1']
plt.figure(figsize=(20,10))
# Graficar las columnas seleccionadas
for c in cols:
    plt.plot(dfr['OBCUpTime'], dfr[c], label=c)

# Añadir etiquetas y leyenda
plt.xlabel('OBCUpTime')
plt.ylabel('temps')
plt.title('-')
plt.legend()

# Mostrar el gráfico
plt.show()
plt.show()
```
